```python
import requests
import pandas as pd

# API URL and Token
api_url = "https://api.socialverseapp.com/posts/view?page=1&page_size=1000"
headers = {
    "Flic-Token": "flic_6e2d8d25dc29a4ddd382c2383a903cf4a688d1a117f6eb43b35a1e7fadbb84b8"
}

# Fetch data from the API
response = requests.get(api_url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract the 'posts' key if it exists
    if 'posts' in data:
        posts = data['posts']
        # Convert the list of posts to a DataFrame
        df = pd.DataFrame(posts)

        # Normalize nested structures (e.g., 'category', 'baseToken')
        if 'category' in df.columns:
            category_df = pd.json_normalize(df['category'])
            category_df.columns = [f"category_{col}" for col in category_df.columns]
            df = pd.concat([df.drop(columns=['category']), category_df], axis=1)

        if 'baseToken' in df.columns:
            baseToken_df = pd.json_normalize(df['baseToken'])
            baseToken_df.columns = [f"baseToken_{col}" for col in baseToken_df.columns]
            df = pd.concat([df.drop(columns=['baseToken']), baseToken_df], axis=1)

        # Preview the processed DataFrame
        print("DataFrame successfully created and normalized:")
        
        # Save to CSV for further analysis
        df.to_csv("posts_data.csv", index=False)
        print("Data saved to 'posts_data.csv'.")
    else:
        print("Key 'posts' not found in the API response.")
else:
    print(f"Failed to fetch data. Status Code: {response.status_code}, Error: {response.text}")
df
```

## Data Cleaning

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Drop columns that are not useful for analysis
columns_to_drop = ['contract_address', 'chain_id', 'chart_url']
df_cleaned = df.drop(columns=columns_to_drop, errors='ignore')

# Fill missing values in numerical columns
numerical_columns = df_cleaned.select_dtypes(include='number').columns
df_cleaned[numerical_columns] = df_cleaned[numerical_columns].fillna(0)

# Fill missing values in categorical columns with 'Unknown'
categorical_columns = df_cleaned.select_dtypes(include='object').columns
df_cleaned[categorical_columns] = df_cleaned[categorical_columns].fillna('Unknown')

print("Data cleaned successfully.")
df_cleaned.head()
```

## Feature Engineering

```python
# Create derived features for engagement
df_cleaned['engagement_score'] = (
    df_cleaned['view_count'] * 0.5 +
    df_cleaned['upvote_count'] * 1.5 +
    df_cleaned['share_count'] * 2 +
    df_cleaned['comment_count'] * 1
)

# Convert timestamp to a human-readable date
df_cleaned['created_date'] = pd.to_datetime(df_cleaned['created_at'], unit='ms')

# Drop the original 'created_at' column
df_cleaned = df_cleaned.drop(columns=['created_at'])

print("Feature engineering completed.")
df_cleaned.head()
```

## Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler

# Select numerical columns for normalization
numerical_columns = ['view_count', 'upvote_count', 'share_count', 'comment_count', 'engagement_score']
scaler = MinMaxScaler()

# Normalize and update DataFrame
df_cleaned[numerical_columns] = scaler.fit_transform(df_cleaned[numerical_columns])

print("Data normalization completed.")
df_cleaned.head()
```

## Content-Based Filtering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Combine relevant features for text-based similarity
df_cleaned['combined_features'] = (
    df_cleaned['category_name'] + " " +
    df_cleaned['title']
)

# Convert combined features into TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_cleaned['combined_features'])

# Compute cosine similarity matrix
similarity_matrix = cosine_similarity(tfidf_matrix)

# Recommend videos similar to a specific video
video_index = 0  # Example: Index of a video the user viewed
similar_videos = list(enumerate(similarity_matrix[video_index]))
similar_videos = sorted(similar_videos, key=lambda x: x[1], reverse=True)

# Display top 5 similar videos
print("Top 5 similar videos:")
for i in range(1, 6):
    print(df_cleaned.iloc[similar_videos[i][0]])
```

## Collaborative Filtering

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

# Create a user-video interaction matrix
interaction_matrix = df_cleaned.pivot_table(
    index='username', 
    columns='id', 
    values='engagement_score', 
    fill_value=0
)

# Convert to compressed sparse row matrix to maintain an effieciency
sparse_matrix = csr_matrix(interaction_matrix.values)

# Use KNN for finding similar users
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(sparse_matrix)

# Find similar users for a given user upto 5 nearest neighbors
user_index = 0  # Example user index
distances, indices = knn.kneighbors(
    sparse_matrix[user_index],
    n_neighbors=5
)

# Recommend videos based on similar users
similar_users = indices.flatten()
recommended_videos = interaction_matrix.iloc[similar_users].mean().sort_values(ascending=False)

print("Recommended videos:")
print(recommended_videos.head(5))
```

## Hybrid Models

```python
# Example: Weighted combination of content-based and collaborative filtering scores
content_scores = similarity_matrix[video_index]  # Content-based similarity scores
collab_scores = recommended_videos.values  # Collaborative filtering scores

# Normalize scores
content_scores = content_scores / max(content_scores)
collab_scores = collab_scores / max(collab_scores[:len(content_scores)])  # Ensure equal length

# Combine scores with weights
hybrid_scores = 0.6 * content_scores + 0.4 * collab_scores

# Recommend videos based on hybrid scores
recommended_indices = hybrid_scores.argsort()[::-1][:5]
print("Hybrid Recommended Videos:")
for idx in recommended_indices:
    print(df_cleaned.iloc[idx])
```

## Mean Absolute Error (MAE)

```python
from sklearn.metrics import mean_absolute_error

# True values: User's actual engagement scores (could be from user interactions or ratings)
true_values = df_cleaned['engagement_score'].values

# Predicted values: The engagement scores predicted by the hybrid recommendation model
# For simplicity, we assume the hybrid_scores (from the previous block) are our predicted values
predicted_values = hybrid_scores  # This is the array from the hybrid recommendation step

# Calculate Mean Absolute Error
mae = mean_absolute_error(true_values, predicted_values)

print(f"Mean Absolute Error (MAE): {mae}")
```

## Root Mean Square Error (RMSE)

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate Root Mean Square Error
rmse = np.sqrt(mean_squared_error(true_values, predicted_values))

print(f"Root Mean Square Error (RMSE): {rmse}")
```
